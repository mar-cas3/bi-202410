```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Datos.csv')
df2 = pd.read_csv('Datos2.csv')
df.head()
```

```python
df2.shape[0]
```

```python
df=df.drop_duplicates()
df2=df2.drop_duplicates()
```

```python
resta=df2.shape[0]-df.shape[0]
df2 = df2.iloc[:-resta]
resta
```

```python
# Seleccionar columnas específicas de cada DataFrame
df1_selected = df[['UUID', 'diagnosisDeTumoresMalignosCancer', 'anioDiagnosis', 'anioDiagnosis_1',
       'anioEncuesta', 'ANIOS_CUMPLIDOS', 'SEXO', 'lugarNacimiento',
       'etnicidad', 'nivelEducativoBin',
       'vecesPorSemanaDeActividadFisica', 'razonDeNoRealizarActividadFisica',
       'haFumadoRecientemente', 'cigarrillosDiariosNum', 'anioEncuesta_1', 
       'codigoLocalidad', 'estrato',
       'cercaniaAIndustria', 'cercaniaATerminalesDeTransporte',
       'cercaniaABasureros', 'cercaniaACaniosDeAguaResidual',
       'cercaniaDeGasolinearias', 'materialParedesExteriores', 'materialPisos',
       'CodgidoLocalidad', '_CodigoLocalidadNum']]
df2_selected = df2[['diagnosisDeProblemasCardiovasculares',
       'problemasDeContaminacionDelAire', 'problemasDeDisposicionDeBasuras']]

# Unir las columnas seleccionadas
df_complete = pd.concat([df1_selected, df2_selected], axis=1)
df_complete.head()
```
